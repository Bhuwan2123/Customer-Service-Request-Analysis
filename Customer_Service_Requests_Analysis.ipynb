```python
dt.columns = dt.columns.str.replace(' ','_')
dt.columns
```

```python
dt.columns = dt.columns.str.replace('(','')
dt.columns = dt.columns.str.replace(')','')
dt.columns
```

```python
dt.info()
```

ols only work with numeric data and if i want to work with string columns than i have to convert them into dummy variables but most of the columns are object type not int or float and also they are more in count so my features will be increase and calculations may not be possible. so i am taking only int or float variables for ols

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dt['Complaint_Type'] = lb.fit_transform(dt['Complaint_Type'])
```

```python
# filtering the int & float columns
x = list(dt.columns[(dt.dtypes == 'int64')|(dt.dtypes == 'float64')])
x
```

```python
import statsmodels.formula.api as sm
ols = sm.ols('Complaint_Type~'+'+'.join(x),data = dt).fit()
ols.summary()
```

the **`p`** values for **`'Y_Coordinate_State_Plane', 'Latitude', 'X_Coordinate_State_Plane', 'Longitude'`** is more than **`0.05`**.

Therefor i should **`accept`** the **`null hypothesis`**(null hypothesis : column has no impact on Complaint_Type) for **`'Y_Coordinate_State_Plane', 'Latitude', 'X_Coordinate_State_Plane', 'Longitude'`**

Therefor i should drop these columns

```python
dt.shape
```

##7. Perform a Kruskal-Wallis H test


        7.1 Fail to reject H0: All sample distributions are equal
        7.2 Reject H0: One or more sample distributions are not equal

```python
x1 = list(dt.columns[(dt.dtypes == 'int64')|(dt.dtypes == 'float64')])
```

```python
x1
```

```python
len(x1)
```

```python
from scipy.stats import kruskal
statistic, pvalue = kruskal(dt[x1[0]], dt[x1[1]], dt[x1[2]], dt[x1[3]], dt[x1[4]], dt[x1[5]], dt[x1[6]])
print("Test statistic:", statistic)
print("P-value:", pvalue)
```

The null hypothesis (H0): The median is equal across all groups.

p value is less than 0.05

**`Therefor`** i should accept the null hypothesis.

This implies  **`The median is equal across ['Unique_Key','Complaint_Type','Incident_Zip','Created_total_seconds','Closed_total_seconds','Due_total_seconds','Resolution_Action_Updated_total_seconds'].`**
